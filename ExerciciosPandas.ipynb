**Bora praticar!**
  
1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
notas = pd.read_csv('dados/alunos3.csv', sep=';')
notas.iloc[:,-4:-1] = notas.iloc[:,-4:-1].apply(lambda x: x.str.replace(',', '.'))
notas
```

```python
notas.info()
```

```python
notas.iloc[:,-4:-1] = notas.iloc[:,-4:-1].astype(float)
notas.info()
```

```python
notas['Media_provas'] = (notas.iloc[:,-4:].sum(axis=1) / 4)
notas
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_nota = notas['Media_provas'].max()
top1 = notas[notas['Media_provas'] == maior_nota]
top1
```

```python
menor_nota = notas['Media_provas'].min()
top_ultimo = notas[notas['Media_provas'] == menor_nota]
top_ultimo
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
cadastro_alunos = pd.read_excel('dados/cadastro_alunos.xlsx')
cadastro_alunos
```

```python
alunos = cadastro_alunos.merge(notas, how='left', left_on='RA', right_on='RA')
alunos
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
media_macho = alunos['Media_provas'].where(alunos['Sexo'] == 'Masculino').mean()
media_macho
```

```python
media_femea = alunos['Media_provas'].where(alunos['Sexo'] == 'Feminino').mean()
media_femea
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
media_idade = alunos['Idade'].where(alunos['Media_provas'] >= 7).mean()
media_idade
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
media_cidade = alunos.groupby('Cidade').agg({'Media_provas': ['mean']})

print(media_cidade)
maior_media = media_cidade.idxmax()
print(maior_media)
print(media_cidade.max())
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
alunos3 = pd.read_csv('dados/alunos3.csv', usecols=['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4'],sep=';')
alunos3.iloc[:,1:-1] = alunos3.iloc[:,1:-1].apply(lambda x: x.str.replace(',', '.'))
alunos3.iloc[:,-4:-1] = alunos3.iloc[:,-4:-1].astype(float)
alunos3['Media_provas'] = alunos3.iloc[:,-4:].sum(axis=1) / 4
alunos3
```

```python
alunos3.to_excel('dados/alunos3_media.xls')
```
